```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Данные по книгам.xlsx')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = df.groupby(by = 'Ниша').sum()
```

```python
df['ЧГ%'] = ((df['ЧГ_2020'] + df['ЧГ_2019']) / (df['ВСЕ_2019'] + df['ВСЕ_2020'])) * 100
```

```python
df['Остальные%'] = ((df['ЛАБИРИНТ_2020'] + df['ЛАБИРИНТ_2019'] + df['ОЗОН_2019'] + df['ОЗОН_2020'] + df['WB_2019'] + df['WB_2020']) / (df['ВСЕ_2019'] + df['ВСЕ_2020'])) * 100
```

```python
df.head()
```

```python
perc = df.filter(['Ниша','Цена базовая', 'ЧГ%','Остальные%'], axis=1)
```

```python
perc['Ниша'] = df.index
```

```python
pc = perc.copy()
```

```python
pc.to_excel('Проценты.xlsx')
```
